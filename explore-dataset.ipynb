```python
def explore_images_per_class(base_path, path):
    dataset_path = base_path+'/'+path
    face_identities = [d for d in os.listdir(dataset_path) if 'n' in d]
    df = pd.DataFrame(columns=['class', 'images'])
    for i in range(0, len(face_identities)):
        dataset_class_path = dataset_path+'/%s' % face_identities[i]
        images_num = [f for f in os.listdir(dataset_class_path) if '.jpg' in f]
        df.loc[i] = [face_identities[i], len(images_num)]
    return df
```
